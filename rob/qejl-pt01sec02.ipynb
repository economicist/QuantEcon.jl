```julia
function factorial2(n)
    num = 1
    for i in 1:n
        num *= i
    end
    return num
end

factorial2(5)
```

```julia
function binomial_rv(n, p)
    successes = 0
    for i in 1:n
        successes += (randn() < p)
    end
    return successes
end
```

```julia
function est_pi(shots)
    hits = 0
    for i in i:shots
        r = rand(2)
        hits += (r[1]^2 + r[2]^2 <= 1)
    end
    return 4*hits/shots
end
```

```julia
est_pi(1000000)
```

```julia
function coinbet(flips, consec2win, show = false)
    heads = 0
    i = 0
    x = String[]
    while (i < flips) & (heads < consec2win)
        i += 1
        push!(x, rand() < 0.5 ? "heads" : "tails")
        heads = x[i] == "heads" ? heads + 1 : 0
    end
    out = [x, string(i) * " flips", i < flips ? "win" : "lose"]
    return show ? out : out[3]
end
```

```julia
coinbet(100, 5, true)
```

```julia
function ar1_data(T, alpha)
    x = Array(Float64, T+1)
    x[1] = 0
    for t in 2:T+1
        x[t] = alpha * x[t-1] + randn()
    end
    return x
end

plot(ar1_data(200, 0.0), "b-", label=L"AR(1),\ \alpha=0.0")
plot(ar1_data(200, 0.8), "g-", label=L"AR(1),\ \alpha=0.8")
plot(ar1_data(200, 0.98), "r-", label=L"AR(1),\ \alpha=0.98")
legend()
```
